```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = (x-5)**2 +3

print(x,y)
```

```python
plt.style.use('seaborn-v0_8')
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y').set_rotation(0)
plt.show()
```

```python
#Gradient Descent
x = 0
lr = 0.1
#lr = 10

# Plot the original curve
x_vals = np.arange(10)
y_vals = (x_vals-5)**2 + 3
plt.plot(x_vals, y_vals, label='Original Curve')


for i in range(50):
  grad=2*(x-5)
  x = x - lr*grad
  y = (x-5)**2 + 3
  plt.scatter(x,y)
  print(x)

plt.show()
```
