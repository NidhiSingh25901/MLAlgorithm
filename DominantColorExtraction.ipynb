```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from sklearn.cluster import KMeans
```

```python
img = cv2.imread('/content/sample_data/Flower.jpg')

# Convert from BGR to RGB for displaying with matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img.shape)
```

```python
plt.imshow(img)
plt.show()
```

```python
x = img.reshape(-1,3)
print(x.shape)
```

```python
k=4
model = KMeans(n_clusters=k)
model.fit(x)
```

```python
centroids = model.cluster_centers_
```

```python
print(centroids)
```

```python
colors = np.array(centroids, dtype='uint8')
```

```python
print(colors)
```

```python
#Color Patch (100,100,3) -> Red Color
i=1
for color in colors:
  print(color)
  plt.subplot(1,k,i)
  plt.axis("off")
  mat = np.zeros((100,100,3), dtype ='uint8')
  mat[:,:,:] = color
  plt.imshow(mat)
  plt.show()
```

```python
#Basic Segmentation based upon similar color regions
#Task1

print(colors)
```

```python
np.unique(model.labels_)
```

```python
newImg = np.zeros((x.shape), dtype ='uint8')
print(newImg.shape)
```

```python
for i in range(newImg.shape[0]):
  newImg[i] = colors[model.labels_[i]]
```

```python
newImg = newImg.reshape(img.shape)
print(newImg.shape)
```

```python
plt.imshow(newImg)
plt.show()
```
