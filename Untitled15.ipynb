```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data Generate x,y

def generateDataset(m):
    x = np.random.randn(m)*10
    noise = np.random.rand(m) *5
    y = 3*x+1 + noise
    return x,y
```

```python
x,y = generateDataset(100)
```

```python
def plotData(x,y, color='orange', title="Linear Regression Example"):
  plt.scatter(x,y, c= color)
  plt.xlabel("x")
  plt.ylabel("y")
  plt.show()
```

```python
plotData(x,y)
```

```python
#Data Normalization is equal to standardization

def normaliseData(x):
  x = (x-x.mean())/x.std()
  return x
```

```python
x= normaliseData(x)
plotData(x,y)
```

```python
def traintestsplit(x,y, split=0.8):
  m = x.shape[0]
  data = np.zeros((m,2))

  data[:,0] = x
  data[:,1] =y

  np.random.shuffle(data)

  split = int(m*split)

  xtrain = data[:split,0]
  ytrain = data[:split,1]

  xtest = data[split:,0]
  ytest = data[split:,1]

  return xtrain,ytrain,xtest,ytest
```

```python
traintestsplit(x,y)
```

```python
xtrain, ytrain, xtest, ytest = traintestsplit(x,y)
```

```python
plt.scatter(xtrain, ytrain, color='orange')
plt.scatter(xtest, ytest, color='blue')
```

```python
#Hypothesis

def hypothesis(x,theta):
  return theta[0]* theta[1]*x

def error(x,y,theta):
  m = x.shape[0]
  e =0
  for i in range(m):
    y_i = hypothesis(x[i], theta)
    e=e+(y[i] - y_i)**2
  return e/(2*m)

def gradient(x,y,theta):
  m = x.shape[0]
  grad = np.zeros((2,))

  for i in range(m):
    exp = hypothesis(x[i],theta) - y[i]
    grad[0]+=(exp)
    grad[1]+= (exp) * (x[i])
  return grad/m

def train(x,y, learning_rate=0.1):
  theta = np.zeros((2,))

  maxItra =100
  error_list = []

  for i in range(maxItra):
    grad = gradient(x,y,theta)
    error_list.append(error(x,y,theta))
    theta[0] = theta[0] - learning_rate * grad[0]
    theta[1] = theta[1] - learning_rate * grad[1]

  plt.plot(error_list)
  plt.show()

  return theta
```

```python
theta = train(x,y)
```
